```python
import pandas as pd
```

```python
from datasets import load_dataset

# データセットの読み込み
dollyDataset = load_dataset("kunishou/databricks-dolly-15k-ja")
sakuraDataset = load_dataset("saldra/sakura_japanese_dataset")

```

```python
dfDolly = pd.DataFrame(dollyDataset['train'])
```

```python
dfDolly[:5]
```

```python
# # いらないもの
# category
# index
```

```python
dfSakura = pd.DataFrame(sakuraDataset['train'])
```

```python
dfSakura[:5]
```

```python
result = pd.concat([dfDolly, dfSakura], axis=0, join="inner")
```

```python
result[:-5]
```

```python
result=result.reset_index(drop=True)
```

```python
result[:-5]
```
